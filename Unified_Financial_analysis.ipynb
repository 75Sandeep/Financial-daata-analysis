```python
# Sort DataFrame by Market Cap in descending order
df_sorted = df.sort_values(by='Mar Cap - Crore', ascending=False)
```

```python
df_sorted
```

```python
# Select top 10 companies
top_10 = df_sorted.head(10)
```

```python
top_10
```

```python
# Plotting the bar chart
plt.figure(figsize=(15, 6))
plt.bar(top_10['Name'], top_10['Mar Cap - Crore'], color='skyblue')
plt.title("Top 10 Companies by Market Cap")
plt.xlabel('Company Names')
plt.ylabel('Market Cap (Crores)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better fit

plt.show()
```
